```python
import PIL
```

```python
import gradio as gr
```

```python
def is_cat(x):
    return x[0].isupper()
```

```python
learner = load_learner("cat_classifier_model.pkl")
```

```python
PIL.Image.open("beautiful-smooth-haired-red-cat-lies-on-the-sofa-royalty-free-image-1678488026.jpg")
```

```python
learner.predict("beautiful-smooth-haired-red-cat-lies-on-the-sofa-royalty-free-image-1678488026.jpg")
```

```python
categories = ("Dog", "Cat")

def classify_img(img):
    pred, index, probs = learner.predict(img)
    return dict(zip(categories, map(float, probs)))
```

```python
image = gr.inputs.Image(shape=(192, 192))
label = gr.outputs.Label()
examples = ['KOA_Nassau_2697x1517.jpg', 'beautiful-smooth-haired-red-cat-lies-on-the-sofa-royalty-free-image-1678488026.jpg']

intf = gr.Interface(fn=classify_img, inputs=image, outputs=label, examples=examples)
intf.launch(inline=False)
```
